```python
ordinal_mapping = {'Uneducated' : 1, 'School' : 2, 'Graduate' : 3, 'Post Graduate' : 4, 'Ph.D' : 5}
data['Educational Qualifications'] = data['Educational Qualifications'].map(ordinal_mapping)
```

```python
data['Monthly Income'].unique()
```

```python
ordinal_mapping_inc = {'No Income' : 1, 'Below Rs.10000' : 2, '10001 to 25000' : 3, '25001 to 50000' : 4, 'More than 50000' : 5}
data['Monthly Income'] = data['Monthly Income'].map(ordinal_mapping_inc)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columns = ['Gender', 'Marital Status', 'Output', 'Feedback', 'Occupation']
for i in columns:
    data[i] = le.fit_transform(data[i])
```

```python
sns.heatmap(data.corr())
```

```python
mean_income_by_age = data.groupby('Age')['Monthly Income'].mean().reset_index()

# Set the style of the seaborn plots
sns.set(style="whitegrid")

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='Monthly Income', data=mean_income_by_age)
plt.title('Average Monthly Income by Age')
plt.xlabel('Age')
plt.ylabel('Average Monthly Income')
plt.show()
```

```python
x = data.drop(columns = 'Feedback')
y = data.iloc[:, -1]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
decision_tree = DecisionTreeClassifier(random_state=90)
decision_tree.fit(x_train, y_train)
```

```python
y_pred_dt = decision_tree.predict(x_test)
```

```python
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print("Decision Tree Accuracy:", accuracy_dt)
```

```python
random_forest = RandomForestClassifier(random_state=52)
random_forest.fit(x_train, y_train)


y_pred_rf = random_forest.predict(x_test)


accuracy_rf = accuracy_score(y_test, y_pred_rf)
print("Random Forest Accuracy:", accuracy_rf)
```
